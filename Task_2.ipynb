```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Gotem Pumpkins.csv")
df.head()
```

```python
df = df.drop(['Convex_Area', 'Equiv_Diameter', 'Solidity', 'Extent'], axis=1)
df.head()
```

```python
X = df.drop('Class', axis=1).values
Y = df['Class'].values
```

```python
for column_index in range(X.shape[1]):
    column = X[:, column_index]
    min_val = column.min()
    max_val = column.max()
    X[:, column_index] = (column - min_val) / (max_val - min_val)
X = X.astype(float)
```

```python
def descent(X, Y, rate=0.01, it=10000):
    m, n = X.shape
    weights = np.zeros(n)
    bias = 0
    for i in range(it):
        lin = np.dot(X, weights) + bias

        yp = 1 / (1 + np.exp(-lin))

        wg = (1/m) * np.dot(X.T, (yp - Y))
        bg = (1/m) * np.sum(yp - Y)

        weights -= rate * wg
        bias -= rate * bg

    return weights, bias, yp
```

```python
for i in range(len(Y)):
    if (Y[i] == 'Ürgüp Sivrisi' ):
        Y[i] = 0
    else:
        Y[i] = 1
Y = Y.astype(float)

weights, bias, yp = descent(X, Y)
print(weights)
print(bias)
```

```python
def pred(X, weights, bias):
    lin = np.dot(X, weights) + bias
    yp = 1 / (1 + np.exp(-lin))
    return [1 if i > 0.5 else 0 for i in yp]
```

```python
y_p = pred(X, weights, bias)
train_accuracy = np.mean(y_p == Y) * 100
print(train_accuracy)
```
